```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import unicodedata
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Carregar os Dados
url_ordens_servico = 'https://raw.githubusercontent.com/Martim35/DECD/main/15-ordens-de-servico.csv'
url_consumos_mensais = 'https://raw.githubusercontent.com/Martim35/DECD/main/3-consumos-faturados-por-municipio-ultimos-10-anos.csv'
url_contadores_energia = 'https://raw.githubusercontent.com/Martim35/DECD/main/21-contadores-de-energia.csv'

ordens_servico = pd.read_csv(url_ordens_servico, delimiter=';')
consumos_mensais = pd.read_csv(url_consumos_mensais, delimiter=';')
contadores_energia = pd.read_csv(url_contadores_energia, delimiter=';')



# Tratar os dados

def limpar_freguesia(nome):
    # Colocar tudo em minúsculas
    nome = nome.lower()
    # Remover acentos
    nome = ''.join((c for c in unicodedata.normalize('NFD', nome) if unicodedata.category(c) != 'Mn'))
    # Remover 'UF' e 'União de Freguesia'
    nome = nome.replace('uniao das freguesias de ', '').replace('uf ', '')
    return nome.strip()


## Limpar os campos da freguesia
ordens_servico['Freguesia'] = ordens_servico['Freguesia'].apply(limpar_freguesia)
consumos_mensais['Freguesia'] = consumos_mensais['Freguesia'].apply(limpar_freguesia)
contadores_energia['Freguesia'] = contadores_energia['Freguesia'].apply(limpar_freguesia)

## Uniformizar os nomes das colunas das tabelas
ordens_servico.rename(columns={'Ordens Serviço Realizadas (#)': 'Ordens Serviço Realizadas', 'data': 'Data'}, inplace=True)
consumos_mensais.rename(columns={'Mês': 'Mes'}, inplace=True)
contadores_energia.rename(columns={'Mês': 'Mes', 'data': 'Data'}, inplace=True)

## Uniformizar as datas em ambas as tabelas
ordens_servico['Data'] = pd.to_datetime(ordens_servico['Data'])
consumos_mensais['Data'] = pd.to_datetime(consumos_mensais['Data'])
contadores_energia['Data'] = pd.to_datetime(contadores_energia['Data'])

## Criar campos novos
tabela = ordens_servico.pivot_table(index=['CodDistritoConcelhoFreguesia', 'Ano', 'Mes'], columns='Tipo de Serviço', values='Ordens Serviço Realizadas', aggfunc='sum', fill_value=0)
tabela.rename(columns=lambda x: f"Ordens por {x}", inplace=True)
ordens_servico = pd.merge(ordens_servico, tabela, on=['CodDistritoConcelhoFreguesia', 'Ano', 'Mes'], how='left')

tabela = consumos_mensais.pivot_table(index=['CodDistritoConcelhoFreguesia', 'Ano', 'Mes'], columns='Nível de Tensão', values='Energia Ativa (kWh)', aggfunc='sum', fill_value=0)
tabela.rename(columns=lambda x: f"Energia por {x}", inplace=True)
consumos_mensais = pd.merge(consumos_mensais, tabela, on=['CodDistritoConcelhoFreguesia', 'Ano', 'Mes'], how='left')

tabela = contadores_energia.pivot_table(index=['CodDistritoConcelhoFreguesia', 'Ano', 'Mes'], columns='Inclui contador inteligente', values="Número de CPE's", aggfunc='sum', fill_value=0)
tabela.rename(columns=lambda x: f"Energia por Contrato={x}", inplace=True)
contadores_energia = pd.merge(contadores_energia, tabela, on=['CodDistritoConcelhoFreguesia', 'Ano', 'Mes'], how='left')

## Remover dados duplicados
ordens_servico.drop_duplicates(subset=['Data', 'CodDistritoConcelhoFreguesia'], inplace=True)
consumos_mensais.drop_duplicates(subset=['Data', 'CodDistritoConcelhoFreguesia'], inplace=True)
contadores_energia.drop_duplicates(subset=['Data', 'CodDistritoConcelhoFreguesia'], inplace=True)

## Remover colunas não relevantes
ordens_servico.drop(columns=['Tipo de Serviço', 'Ordens Serviço Realizadas'], inplace=True)
consumos_mensais.drop(columns=['CodDistrito', 'CodDistritoConcelho', 'Nível de Tensão', 'Energia Ativa (kWh)'], inplace=True)
contadores_energia.drop(columns=['CodDistrito', 'CodDistritoConcelho', 'Inclui contador inteligente', "Número de CPE's", 'Contrato Ativo'], inplace=True)

### Verificar os campos
print('Campos antes do merge')
print(ordens_servico.info())
print(consumos_mensais.info())
print(contadores_energia.info())
print('------------')


# Merge

## Merge das tabelas por Data e Freguesia
dados = pd.merge(ordens_servico, consumos_mensais, on=['Data', 'CodDistritoConcelhoFreguesia'], how='inner')
dados = pd.merge(dados, contadores_energia, on=['Data', 'CodDistritoConcelhoFreguesia'], how='inner')

## Remover colunas desnecessárias e alterar os nomes
dados.drop_duplicates(inplace=True)
dados.drop(columns=['Ano_y', 'Mes_y', 'Distrito_y', 'Concelho_y', 'Freguesia_y', 'Ano', 'Mes', 'Distrito', 'Concelho', 'Freguesia'], inplace=True)
dados.rename(columns={'Ano_x': 'Ano', 'Mes_x': 'Mes', 'Distrito_x': 'Distrito', 'Concelho_x': 'Concelho', 'Freguesia_x': 'Freguesia'}, inplace=True)

# Agrupar os dados por Concelho e Mês
dados = dados.groupby(['Concelho', 'Ano', 'Mes']).agg({
    'Ordens por Configuração de EMI': 'sum',
    'Ordens por Outros': 'sum',
    'Ordens por REMOTE_ICP_CONFIGURATION': 'sum',
    'Ordens por Redução de Potência Contratada': 'sum',
    'Ordens por Reposição de Potência Contratada': 'sum',
    'Ordens por Servico Alteração Contractual': 'sum',
    'Ordens por Serviço de Corte e Re-Ligação': 'sum',
    'Energia por Baixa Tensão': 'sum',
    'Energia por Muito Alta, Alta e Média Tensões': 'sum',
    'Energia por Contrato=SIM': 'sum',
    'Energia por Contrato=NÃO': 'sum',
}).reset_index()

## Criar a variável alvo
dados['Total de Ordens de Serviço'] = dados.filter(like='Ordens por').sum(axis=1)
dados.drop(columns=['Ordens por Configuração de EMI', 'Ordens por Outros', 'Ordens por REMOTE_ICP_CONFIGURATION', 'Ordens por Redução de Potência Contratada', 'Ordens por Reposição de Potência Contratada', 'Ordens por Servico Alteração Contractual', 'Ordens por Serviço de Corte e Re-Ligação'], inplace=True)

# Visualizar os dados agrupados
print('Dados agrupados')
print(dados.head())
print(dados.info())
print('------------')

# Valores nulos
null_counts = dados.isnull().sum()
print("Valores nulos:\n", null_counts)
print('------------')

print('Correlações')
# Correlações

## Variáveis relevantes e a variável alvo
variavel_alvo = 'Total de Ordens de Serviço'
variaveis= [
    'Energia por Baixa Tensão',
    'Energia por Muito Alta, Alta e Média Tensões',
    'Energia por Contrato=SIM',
    'Energia por Contrato=NÃO',
    'Total de Ordens de Serviço'
]

dados_variaveis = dados[variaveis]

## Correlação entre as variáveis relevantes e a variável alvo
correlacao = dados_variaveis.corr()
correlacao = correlacao[[variavel_alvo]].sort_values(by=variavel_alvo, ascending=False)

## Heatmap das correlações com a variável alvo
plt.figure(figsize=(2, 3))
sns.heatmap(correlacao, annot=True, cmap='coolwarm', vmin=-1, vmax=1, fmt=".2f", cbar_kws={'shrink': .8})
plt.title(f'Heatmap de Correlações com Total de Ordens de Serviço')
plt.show()

print('------------')
print('Trasnformação + Normalização')
# Trasnformação + Normalização

## Visualização dos dados iniciais
plt.figure(figsize=(6, 4))
sns.boxplot(data=dados_variaveis.drop(columns=['Total de Ordens de Serviço']))
plt.xticks(rotation=90)
plt.title('Boxplot das Variáveis em Análise')
plt.show()

plt.figure(figsize=(6, 4))
dados_variaveis.drop(columns=['Total de Ordens de Serviço']).hist(bins=50, figsize=(15, 10))
plt.suptitle('Histograma das Variáveis em Análise')
plt.show()

## Transformação Logarítmica
dados_log = dados_variaveis.copy()
for coluna in dados_variaveis.columns[:-1]:
    dados_log[coluna] = np.log1p(dados_log[coluna])

## Visualização após transformação logarítmica
plt.figure(figsize=(6, 4))
sns.boxplot(data=dados_log.drop(columns=['Total de Ordens de Serviço']))
plt.xticks(rotation=90)
plt.title('Boxplot das Variáveis com Transformação Logarítmica')
plt.show()

plt.figure(figsize=(6, 4))
dados_log.drop(columns=['Total de Ordens de Serviço']).hist(bins=50, figsize=(15, 10))
plt.suptitle('Histograma das Variáveis com Transformação Logarítmica')
plt.show()

## PCA sem transformação
scaler = StandardScaler()
dados_normalizados = scaler.fit_transform(dados_variaveis.drop(columns=[variavel_alvo]))

pca = PCA()
componentes_principais = pca.fit_transform(dados_normalizados)

# Gráfico da variância acumulada
plt.figure(figsize=(8, 4))
plt.plot(np.cumsum(pca.explained_variance_ratio_), marker='o', linestyle='--')
plt.xlabel('Número de Componentes Principais')
plt.ylabel('Variância Acumulada')
plt.title('Variância Acumulada por PCA')
plt.grid()
plt.show()

## Número de componentes com base na variância acumulada
n_componentes = 2

## PCA sem transformação
pca = PCA(n_components=n_componentes)
componentes_principais = pca.fit_transform(dados_normalizados)

plt.figure(figsize=(6, 4))
sns.scatterplot(x=componentes_principais[:, 0], y=componentes_principais[:, 1], hue=dados_variaveis[variavel_alvo], palette='coolwarm')
plt.title('PCA Sem Transformação')
plt.show()

## PCA com transformação logarítmica
dados_normalizados_log = scaler.fit_transform(dados_log.drop(columns=[variavel_alvo]))

pca_log = PCA(n_components=2)
componentes_principais_log = pca_log.fit_transform(dados_normalizados_log)

plt.figure(figsize=(6, 4))
sns.scatterplot(x=componentes_principais_log[:, 0], y=componentes_principais_log[:, 1], hue=dados_log[variavel_alvo], palette='coolwarm')
plt.title('PCA com Transformação Logarítmica')
plt.show()


print('Outliers')
# Outliers

## Função para remover outliers
def remover_outliers(df):
    coluna = 'Energia por Muito Alta, Alta e Média Tensões'
    Q1 = df[coluna].quantile(0.25)
    IQR = Q1 - df[coluna].quantile(0.75)
    limite_inferior = Q1 - 1.5 * IQR
    df = df[df[coluna] >= limite_inferior]

    return df


dados_log_sem_outliers = remover_outliers(dados_log)

plt.figure(figsize=(6, 4))
sns.boxplot(data=dados_log_sem_outliers.drop(columns=['Total de Ordens de Serviço']))
plt.xticks(rotation=90)
plt.title('Boxplot das Variáveis Log sem Outliers')
plt.show()

## PCA com transformação logarítmica sem outliers
dados_log_sem_outliers_normalizados = scaler.fit_transform(dados_log_sem_outliers)

pca_log = PCA(n_components=2)
componentes_principais_log_sem_outliers = pca_log.fit_transform(dados_log_sem_outliers_normalizados)

# Deu erro sem
dados_log_sem_outliers.reset_index(drop=True, inplace=True)

plt.figure(figsize=(6, 4))
sns.scatterplot(x=componentes_principais_log_sem_outliers[:, 0], y=componentes_principais_log_sem_outliers[:, 1], hue=dados_log_sem_outliers[variavel_alvo], palette='coolwarm')
plt.title('PCA com Transformação Logarítmica sem Outliers')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.show()


# KMeans
def avaliar_kmeans(n_clusters, dados):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    clusters = kmeans.fit_predict(dados)

    dados['Cluster'] = clusters

    # Gráficos dos Clusters
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=dados.iloc[:, 0], y=dados.iloc[:, 1], hue=clusters, palette='viridis')
    plt.title(f'Clusters do KMeans com {n_clusters} Clusters')
    plt.xlabel('Componente Principal 1')
    plt.ylabel('Componente Principal 2')
    plt.show()

    # Métricas de avaliação
    silhouette = silhouette_score(dados.iloc[:, :-1], clusters)
    davies_bouldin = davies_bouldin_score(dados.iloc[:, :-1], clusters)
    calinski_harabasz = calinski_harabasz_score(dados.iloc[:, :-1], clusters)

    return silhouette, davies_bouldin, calinski_harabasz

## Comparar resultados de KMeans com diferentes números de clusters
resultados = {}
for n_clusters in range(2, 6):
    dados_para_kmeans = pd.DataFrame(componentes_principais_log_sem_outliers, columns=['PCA1', 'PCA2'])
    silhouette, davies_bouldin, calinski_harabasz = avaliar_kmeans(n_clusters, dados_para_kmeans)
    resultados[n_clusters] = {
        'Silhouette': silhouette,
        'Davies-Bouldin': davies_bouldin,
        'Calinski-Harabasz': calinski_harabasz
    }

## Resultados
for n_clusters, metrics in resultados.items():
    print(f'Número de Clusters: {n_clusters}')
    print(f"Silhouette Score: {metrics['Silhouette']:.4f}")
    print(f"Davies-Bouldin Score: {metrics['Davies-Bouldin']:.4f}")
    print(f"Calinski-Harabasz Score: {metrics['Calinski-Harabasz']:.4f}")
    print('-------------')


# Clustering Hierárquico
Z = linkage(componentes_principais_log_sem_outliers, method='ward')

## Dendrograma
plt.figure(figsize=(10, 6))
dendrogram(Z, truncate_mode='level', p=5)
plt.title('Dendrograma para Clustering Hierárquico')
plt.xlabel('Índice da Amostra')
plt.ylabel('Distância')
plt.show()

## Função para cortar o dendrograma em diferentes níveis e avaliar clusters
def avaliar_clusters_hierarquicos(Z, max_distancias):
    for max_d in max_distancias:
        clusters_hierarquico = fcluster(Z, max_d, criterion='distance')

        silhouette_avg = silhouette_score(componentes_principais_log_sem_outliers, clusters_hierarquico)
        db_score = davies_bouldin_score(componentes_principais_log_sem_outliers, clusters_hierarquico)
        ch_score = calinski_harabasz_score(componentes_principais_log_sem_outliers, clusters_hierarquico)

        plt.figure(figsize=(10, 6))
        sns.scatterplot(x=componentes_principais_log_sem_outliers[:, 0], y=componentes_principais_log_sem_outliers[:, 1], hue=clusters_hierarquico, palette='viridis')
        plt.title(f'Clusters Hierárquicos com Corte a {max_d}')
        plt.xlabel('Componente Principal 1')
        plt.ylabel('Componente Principal 2')
        plt.show()

        print(f'Corte a Distância: {max_d}')
        print(f'Silhouette Score: {silhouette_avg:.4f}')
        print(f'Davies-Bouldin Score: {db_score:.4f}')
        print(f'Calinski-Harabasz Score: {ch_score:.4f}')
        print('--------------')

## Avaliar clusters com diferentes níveis de corte no dendrograma
max_distancias = [5, 10, 15, 20]
avaliar_clusters_hierarquicos(Z, max_distancias)



## DBSCAN com diferentes parâmetros
eps_values = [0.2, 0.3, 0.4, 0.5]
min_samples_values = [5, 10, 15]

def avaliar_dbscan(componentes_principais, eps_values, min_samples_values):
    for eps in eps_values:
        for min_samples in min_samples_values:
            dbscan = DBSCAN(eps=eps, min_samples=min_samples)
            clusters_dbscan = dbscan.fit_predict(componentes_principais)

            # Verificar se há clusters formados
            if len(set(clusters_dbscan)) > 1:
                silhouette_avg = silhouette_score(componentes_principais, clusters_dbscan)
                db_score = davies_bouldin_score(componentes_principais, clusters_dbscan)
                ch_score = calinski_harabasz_score(componentes_principais, clusters_dbscan)

                plt.figure(figsize=(10, 6))
                sns.scatterplot(x=componentes_principais[:, 0], y=componentes_principais[:, 1], hue=clusters_dbscan, palette='viridis', legend='full')
                plt.title(f'Clusters DBSCAN com eps={eps} e min_samples={min_samples}')
                plt.xlabel('Componente Principal 1')
                plt.ylabel('Componente Principal 2')
                plt.show()

                print(f'eps: {eps}, min_samples: {min_samples}')
                print(f'Silhouette Score: {silhouette_avg:.4f}')
                print(f'Davies-Bouldin Score: {db_score:.4f}')
                print(f'Calinski-Harabasz Score: {ch_score:.4f}')
                print('-----------')


# Avaliar DBSCAN
avaliar_dbscan(componentes_principais_log_sem_outliers, eps_values, min_samples_values)


'''
print('Dados ')
print(dados.head())
print(dados.info())
print('------------')
print('Dados 2')
print(dados_log_sem_outliers.head())
print(dados_log_sem_outliers.info())
print('------------')
'''

#Parte 2

## Adicionar o ANo
dados_log_sem_outliers['Ano'] = dados['Ano']

## Componentes principais
X = componentes_principais_log_sem_outliers
y = dados_log_sem_outliers[variavel_alvo].values

## Dividir os dados em treino e teste (usando dados de 2024 como teste)
X_train = X[dados_log_sem_outliers['Ano'] != 2024]
y_train = y[dados_log_sem_outliers['Ano'] != 2024]
X_test = X[dados_log_sem_outliers['Ano'] == 2024]
y_test = y[dados_log_sem_outliers['Ano'] == 2024]

## Normalizar os dados de novo
scaler = StandardScaler()
X_train_normalizado = scaler.fit_transform(X_train)
X_test_normalizado = scaler.transform(X_test)

# RandomForestRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

rf = RandomForestRegressor(random_state=42)

## Hiperparâmetros
param_grid_rf = {
    'n_estimators': [100, 200],
    'max_depth': [10, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}

## Validação cruzada
grid_search_rf = GridSearchCV(estimator=rf, param_grid=param_grid_rf, cv=3, scoring='r2', n_jobs=-1, verbose=2)
grid_search_rf.fit(X_train_normalizado, y_train)

best_rf = grid_search_rf.best_estimator_

## previsões no conjunto de teste
y_pred_rf = best_rf.predict(X_test_normalizado)

## Avaliação do modelo
mse_rf = mean_squared_error(y_test, y_pred_rf)
mae_rf = mean_absolute_error(y_test, y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)

print(f'Mean Squared Error (Random Forest): {mse_rf}')
print(f'Mean Absolute Error (Random Forest): {mae_rf}')
print(f'R^2 Score (Random Forest): {r2_rf}')

# Exibir os melhores hiperparâmetros
print(f'Melhores Hiperparâmetros (Random Forest): {grid_search_rf.best_params_}')

# Visualização das previsões
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred_rf, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.xlabel('Valores Reais')
plt.ylabel('Previsões')
plt.title('Previsões vs Valores Reais (Random Forest)')
plt.show()

# Regressão Linear
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train_normalizado, y_train)
y_pred_lr = lr.predict(X_test_normalizado)

## Avaliação do modelo
mse_lr = mean_squared_error(y_test, y_pred_lr)
mae_lr = mean_absolute_error(y_test, y_pred_lr)
r2_lr = r2_score(y_test, y_pred_lr)

print(f'Mean Squared Error (Linear Regression): {mse_lr}')
print(f'Mean Absolute Error (Linear Regression): {mae_lr}')
print(f'R^2 Score (Linear Regression): {r2_lr}')

## Visualização
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred_lr, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.xlabel('Valores Reais')
plt.ylabel('Previsões')
plt.title('Previsões vs Valores Reais (Linear Regression)')
plt.show()

# DecisionTreeRegressor
from sklearn.tree import DecisionTreeRegressor

dt = DecisionTreeRegressor(random_state=42)

# Definir a grade de hiperparâmetros
param_grid_dt = {
    'max_depth': [10, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}

## Validação cruzada
grid_search_dt = GridSearchCV(estimator=dt, param_grid=param_grid_dt, cv=3, scoring='r2', n_jobs=-1, verbose=2)
grid_search_dt.fit(X_train_normalizado, y_train)

best_dt = grid_search_dt.best_estimator_

## Previsões
y_pred_dt = best_dt.predict(X_test_normalizado)

## Avaliação do modelo
mse_dt = mean_squared_error(y_test, y_pred_dt)
mae_dt = mean_absolute_error(y_test, y_pred_dt)
r2_dt = r2_score(y_test, y_pred_dt)

print(f'Mean Squared Error (Decision Tree): {mse_dt}')
print(f'Mean Absolute Error (Decision Tree): {mae_dt}')
print(f'R^2 Score (Decision Tree): {r2_dt}')

# Exibir os melhores hiperparâmetros
print(f'Melhores Hiperparâmetros (Decision Tree): {grid_search_dt.best_params_}')

## Previsões
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred_dt, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.xlabel('Valores Reais')
plt.ylabel('Previsões')
plt.title('Previsões vs Valores Reais (Decision Tree)')
plt.show()
```
